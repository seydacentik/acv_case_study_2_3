```python
hamamozu_musteri = df_tahakkuk["sozlesme_hesap_no"].nunique()
gumushacikoy_musteri = df_tahakkuk_1["sozlesme_hesap_no"].nunique()
goynucek_musteri = df_tahakkuk_2["sozlesme_hesap_no"].nunique()

print("Hamamözü:", hamamozu_musteri)
print("Gümüşhacıköy:", gumushacikoy_musteri)
print("Göynücek:", goynucek_musteri)
```

```python
print("Hamamözü:", df_tahakkuk.shape[0])
print("Gümüşhacıköy:", df_tahakkuk_1.shape[0])
print("Göynücek:", df_tahakkuk_2.shape[0])
```

```python
df_tahakkuk["ilce"] = "Hamamözü"
df_tahakkuk_1["ilce"] = "Gümüşhacıköy"
df_tahakkuk_2["ilce"] = "Göynücek"
```

```python
df_tahakkuk_all = pd.concat(
    [df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2],
    ignore_index=True
)
```

```python
print("Birleşik Tahakkuk:", df_tahakkuk_all.shape[0])
df_tahakkuk.shape[0] + df_tahakkuk_1.shape[0] + df_tahakkuk_2.shape[0]
```

```python
df_tahakkuk_all["ilce"].value_counts()
```

```python
df_tahakkuk_all["kwh"] = pd.to_numeric(
    df_tahakkuk_all["kwh"],
    errors="coerce"
)
```

```python
df_tahakkuk_all["kwh"].isna().sum()
```

```python
df_tahakkuk_all[df_tahakkuk_all["kwh"] < 0].shape
```

```python
df_kwh_valid = df_tahakkuk_all[
    (df_tahakkuk_all["kwh"].notna()) &
    (df_tahakkuk_all["kwh"] >= 0)
]

df_kwh_valid.shape
```

```python
df_kwh_valid["kwh"].describe()
```

```python
tablo = (
    df_kwh_valid
    .groupby("hesap_sinifi")["kwh"]
    .agg(
        ortalama_kwh="mean",
        medyan_kwh="median",
        std_kwh="std",
        kayit_sayisi="count"
    )
    .reset_index()
)

tablo
```
